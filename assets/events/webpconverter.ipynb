```python
def convert_to_webp(image_filename, source_folder, destination_folder):
    image_path = os.path.join(source_folder, image_filename)
    destination_path = os.path.join(destination_folder, image_filename[:image_filename.find(".")] + ".webp") 
    
    if not os.path.exists(destination_path):
        print(source_folder,image_filename)
        print(image_filename)
        image = Image.open(image_path)  # Open image
        image.save(destination_path, format="webp")  # Convert image to webp
    
    return destination_path
```

```python
for event_folder in [x for x in os.listdir(".") if "." not in x and x != "output"]:
    path = os.path.join(".", event_folder)
    destination = os.path.join(".","output", event_folder)

    if not os.path.exists(destination):
        os.mkdir(destination)

    for image in [x for x in os.listdir(event_folder) if x != ".DS_Store"]:
        convert_to_webp(image, path, destination)
        
```
